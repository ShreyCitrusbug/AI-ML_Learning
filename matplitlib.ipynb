```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
historical_prices = pd.read_csv("HistoricalPrices.csv")
```

```python
historical_prices = historical_prices.rename(
    columns={' Open': 'Open', ' High': 'High', ' Low': 'Low', ' Close': 'Close'})
historical_prices.head()
```

```python
historical_prices['Date'] = pd.to_datetime(historical_prices['Date'])
historical_prices = historical_prices.sort_values(by='Date')
```

```python
plt.plot(historical_prices['Date'],historical_prices['Close'],label="Close")
plt.legend()
plt.show()
```

```python
# Bar chart
from calendar import month_name
historical_prices['Month'] = pd.Categorical(historical_prices['Date'].dt.month_name())
historical_prices.head()
plt.bar(historical_prices['Month'],historical_prices['Date'])
plt.show()
```

```python
#Horizontal Bar Chart
plt.barh(historical_prices['Date'],historical_prices['Month'])
```

```python
# Scatter Plot
plt.scatter(historical_prices['Date'],historical_prices['Close'])
plt.title("Open vs. Close Price")
plt.xlabel("Date")
plt.ylabel("Close Price")
```
