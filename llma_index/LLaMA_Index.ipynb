Load data and build an index

```python
import os
import openai
```

```python
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader, TreeIndex

openai.api_key = os.getenv("OPEN_API_KEY")
```

```python
# document = SimpleDirectoryReader("data").load_data()
# index = VectorStoreIndex.from_documents(documents=document)
resume = SimpleDirectoryReader("pdf").load_data()
new_index = TreeIndex.from_documents(resume)
```

```python
new_index
```

Query your data

```python
query_engine = new_index.as_query_engine()
response = query_engine.query("When did Abid graduated?")
print(response)
```
