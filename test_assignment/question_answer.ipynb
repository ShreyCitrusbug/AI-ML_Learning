```python
import fitz
import json
import re
from datetime import datetime
```

```python
datetime.now().strftime("%Y-%m-%d")
```

```python
# Function to extract text from PDF
def extract_text_from_pdf(pdf_path):
    text = ""
    with fitz.open(pdf_path) as doc:
        for page_num in range(len(doc)):
            page = doc.load_page(page_num)
            text += page.get_text()
    return text
```

```python
# Function to parse questions and options
def parse_questions(text):
    questions = []
    question_pattern = re.compile(
        r"Q\d+\.(.*?)(?=Q\d+\.|$)", re.S
    )  # Regex to match questions
    option_pattern = re.compile(r"[a-dA-D]\)\s.*")  # Regex to match options

    matches = question_pattern.findall(text)
    print(matches, "matches")
    for match in matches:
        print(match, "match")
        lines = match.strip().split("\n")
        question_text = lines[0].strip()
        print(question_text, "question_text")
        options = [line.strip() for line in lines[1:] if option_pattern.match(line)]
        questions.append({"question": question_text, "options": options})

    return questions
```

```python
def convert_pdf(pdf_path: str, json_path: str) -> dict:
    text = extract_text_from_pdf(pdf_path=pdf_path)
    questions = parse_questions(text=text)
    print(questions, "questions")
    with open(json_path, "w") as json_file:
        json.dump(questions, json_file, indent=4)
```

```python
pdf_path = "Draft-OASIS-E All Items_03122020.pdf"
json_path = f"{datetime.now().strftime('%Y-%m-%d')}.json"
```

```python
convert_pdf(pdf_path, json_path)
```
